```python
df3 = pd.read_csv("data_3.csv", index_col=0, encoding='utf-8')
df3.head()
```

```python
df3['built_up'] = df3['built_up'].map(lambda x: x.lstrip('From  Up to ').rstrip(' sq. ft.'))
df3
```

```python
df3['built_up'] = [np.mean(pd.to_numeric(x.strip('- ').split('-'), errors='coerce')) for x in df3.built_up]
df3
```

```python
df3 = df3.dropna()
df3
```

```python
df3.hist(column='built_up')
```

```python
df3['bedrooms'] = df3['bedrooms'].str.replace('Studio', '1')        
df3['bedrooms'] = df3['bedrooms'].str.rstrip('+- ').apply(pd.eval)
df3['bedrooms'] = abs(df3['bedrooms'])
df3
```

```python
df3['monthly_installment'] = df3['monthly_installment'].str.replace(',', '')   
df3['monthly_installment'] = pd.to_numeric(df3['monthly_installment'])
df3
```

```python
df3.type.value_counts().plot(kind='barh', figsize=(10, 10))
```

```python
df3.loc[(df3.type.str.match(pat = '(Bungalow)|(Villa)') ),'type']='Bungalow/Villa'
df3.loc[(df3.type.str.match(pat = '(Apartment)|(Condominium)|(Serviced Residence)|(Flat)') ),'type']='Apartment/Condo/Service Residence'
df3.loc[(df3.type.str.contains('|'.join(['Semi-Detached','Semi Detached']), flags=re.IGNORECASE, regex=True) ),'type']='Semi-Detached House'
df3.loc[(df3.type.str.contains('|'.join(['Terrace/Link','Townhouse', 'Cluster', 'Superlink']), flags=re.IGNORECASE, regex=True) ),'type']='Terrace/Link House'
df3.loc[(df3.type.str.match(pat = '(Residential Land)') ),'type']='Residential Land'
```

```python
Q1 = df3.quantile(0.25)
Q3 = df3.quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
df3 = df3[~((df3 < (Q1 - 1.5 * IQR)) |(df3 > (Q3 + 1.5 * IQR))).any(axis=1)]
df3.shape
```

```python
cleanup_nums = {"type":{"Bungalow/Villa": 1, 
                        "Apartment/Condo/Service Residence": 2,
                        "Semi-Detached House": 3,
                        "Terrace/Link House": 4,
                        "Residential Land": 5},
                "land_title":{"Industrial": 1, 
                              "Commercial": 2, 
                              "Unknown": 3, 
                              "Residential": 4},
                "tenure":{"Malay Reserved Land":1,
                          "Lease Tenancy":2,
                          "Leasehold":3,
                          "Unknown":4,
                          "Freehold":5},
                "furnishing":{"Fully Furnished":1,
                              "Unfurnished":2,
                              "Unknown":3,
                              "Partly Furnished":4}
               }

df3.replace(cleanup_nums, inplace=True)
df3.tail()
```
